### **Data Reading**

```python
df=spark.readStream.format('cloudFiles')\
    .option('cloudFiles.format','parquet')\
        .option('cloudFiles.schemaLocation',f"abfss://bronze2026@storagespotifyproject20.dfs.core.windows.net/checkpoint_{p_file_name}")\
            .load(f"abfss://source@storagespotifyproject20.dfs.core.windows.net/{p_file_name}")
```

## Data Writing

```python
df.writeStream.format('parquet').\
    outputMode('append').\
        option('checkpointLocation',f"abfss://bronze2026@storagespotifyproject20.dfs.core.windows.net/checkpoint_{p_file_name}").option('path',f"abfss://bronze2026@storagespotifyproject20.dfs.core.windows.net/{p_file_name}").trigger(once=True).start()
```
